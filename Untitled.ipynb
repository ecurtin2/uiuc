```python
# -*- coding: utf-8 -*-
"""
Created on Mon Oct 24 13:06:57 2016

@author: Gloria
"""
%matplotlib inline

#==============================================================================
# #Load Necessary Packages
#==============================================================================
import matplotlib.pyplot as plt
import numpy as np
from scipy import special
import scipy.special as sp
from math import pi
from math import factorial

#==============================================================================
# #Import PM7 Data 
#==============================================================================
#PM7data = np.genfromtxt("C:\\Users\Glo\Desktop\curve data atu.csv",delimiter = ',')
##print (PM7data)
#OHdistance = PM7data[:,0]
#Energy = PM7data[:,1]
##print (OHdistance, Energy)
##plt.title('PM7 Constrained Optimization of pHPP')
##plt.xlabel('O-H Distance(Bohr)')
##plt.ylabel('Energy (hartrees)')
#==============================================================================
# #Define xmin (where the plot starts), xketo (where the keto well starts), xmax (where the plot ends)
#==============================================================================
xmin=0
xketo=2.50
xmax=10.0
#==============================================================================
# #Define Potential Function U
#==============================================================================
def Potential(x):
#First the Enol well which is fit to a quadratic polynomial 
    if xmin <= x <= xketo:
        return 0.5*(0.35)*(x-Enolminx)**2 + 0.006
#Then the Keto well whch is fit to a quartic polynomial 
    elif xketo < x <= xmax:
        return (0.00117331*(x**4)-0.02357650*(x**3)+0.18374222*(x**2)-0.66274089*(x)+0.93286982)
    else:
        return 0
    
Enolminx = 1.853
Ketominx = 5.519       
#Defining force constants and frequencies 
kenol = 0.35
wenol = (kenol)**0.5
kketo = 0.18374222
wketo = (kketo)**0.5
#==============================================================================
#Plot both the piecewise function defined potential (blue line) and the PM7 results (points)
#==============================================================================
#This has to be done before the function can be plotted
x = np.linspace(xmin, xmax, 100)     
Ufunc = np.vectorize(Potential) 
U = Ufunc(x)
#Plotting the potential with the PM7 points overlayed 
plt.figure(1) 
plt.title("Double Well Potential ") 
plt.xlabel("O-H distance (bohr)") 
plt.ylabel("U(O-H) (hartrees)")
plt.xlim(xmin,xmax-1)
plt.ylim(-0.05,0.40)
plt.plot(x,U,'k--')
#plt.plot(OHdistance,Energy,'ro')
plt.show()
#==============================================================================
# #Define basis in Enol Well
#==============================================================================
#Number of basis functions on the enol well

```

```python
import scipy.special as sp
import math

def HO_wavefunction(x, n, x0, omega, m=1):  # m=1 like this is a default value
                                            # can be called with or without m. If not specified, m=1.
        
    N = 1.0 / math.sqrt(2**n * math.factorial(n)) * ((m * omega)/(math.pi))**(0.25) # Normaliziation
    y = np.sqrt(m * omega) * (x - x0)                  # redefine variables to make it cleaner
    return N * np.exp(-y * y / 2.0) * sp.hermite(n)(y) # N * gausssian * hermite

def enol_basis(x, n):
    return HO_wavefunction(x, n, x0=Enolminx, omega=wenol)

def keto_basis(x, n):
    return HO_wavefunction(x, n, x0=Ketominx, omega=wketo)


# Starting figure 1
fig = plt.figure()
x = np.linspace(xmin, xmax, 100)  # creates evenly spaced grid from xmin to xmax, with 100 points
plt.plot(x, HO_wavefunction(x, n=1, x0=5, omega=1))
plt.show()


# Starting figure 2
fig = plt.figure()

V_array = np.asarray([Potential(x_i) for x_i in x])
scale = 0.15

plt.title("Double Well Potential ") 
plt.xlabel("O-H distance (bohr)") 
plt.ylabel("U(O-H) (hartrees)")
plt.xlim(xmin,xmax-1)
plt.ylim(-0.05,0.40)
plt.plot(x, V_array, 'k--')           # plot potential
plt.plot(x, scale * enol_basis(x, 0)) # plot basis, the scale is for visualization
plt.plot(x, scale * keto_basis(x, 0))
plt.show()
```
